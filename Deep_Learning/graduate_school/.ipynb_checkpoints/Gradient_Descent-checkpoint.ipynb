### 주어진 데이터에 가장 잘 부합하는 모델을 찾는 문제   
- w0, w1, w2를 GDM으로 결정하기 

#### 기본 값 정의

#### Error 함수 구하기 

#### w0, w1, w2에 대한 편미분값 구하기

### 구현

```python
def gradient_descent(lr, epochs):
    w0 = 1 
    w1 = 1
    w2 = 1 
    
    for epoch in range(epochs):
        # error 함수
        error = 18*(w2**2) + 6*(w1**2) + 4*(w0**2) + 20*w1*w2 + 12*w0*w2 + 8*w0*w1 - 14*w2 - 10*w1 - 8*w0 + 6 
        
        # w0, w1, w2에 대한 편미분 값 
        w0_d = 8*w0 + 12*w2 + 8*w1 -8 
        w1_d = 8*w0 + 12*w1 + 20*w2 - 10
        w2_d = 36*w2 + 20*w1 + 12*w0 -14 
        
        # error가 특정 수치보다 낮으면 반복문 종료 
        if error <= 0.001: 
            break
        
        # w0, w1, w2 가중치 update 
        w0 = w0 - lr*w0_d
        w1 = w1 - lr*w1_d
        w2 = w2 - lr*w2_d
        
        # 100번째마다 학습 결과 출력
        if epoch in range(0, epochs, 100):   
            print(f"epoch: {epoch}, w0: {w0}, w1: {w1}, w2: {w2}, error: {error}")
```

```python
learning_rate = 0.01 
epochs = 10000
```

```python
gradient_descent(learning_rate, epochs)
```
